# 2. Apply Otsu thresholding on original images

We want to perform Otsu thresholing on the dataset **N2DL-Hela**. The dataset consists of 4 images. Otsu is not a classic clustering method, but a method that finds an optimal threshold value in a grayscale image so that it can be divided into two classes. The optimal threshold value is the one where the between-class variance is maximum and the within-class variance is minimum. When the optimum threshold value is found, all pixels with an intensity value below the threshold value are set to 0 (black), and all pixels above the threshold value are set to 255 (white). The result is a binary image.

Our goal is therefore to use Otsu to generate a binary image in which the cells are clearly highlighted in contrast to the background.
These results are then compared with the ground truth mask using the dice score to check how well the segmentation with Otsu worked.

```python
Otsu_t13 = otsu_thresholding(t13)
Otsu_t52 = otsu_thresholding(t52)
Otsu_t75 = otsu_thresholding(t75)
Otsu_t79 = otsu_thresholding(t79)

display_images(Otsu_t13, "Otsu_t13")
display_images(Otsu_t52, "Otsu_t52")
display_images(Otsu_t75, "Otsu_t75")
display_images(Otsu_t79, "Otsu_t79")
```
