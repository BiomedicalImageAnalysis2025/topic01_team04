```python
from Functions.sklearn_segmentation import skcluster_image, save_image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
import os

image_paths = [
    'Images/Datenvorbereitung_Images/CellNuclei/HSV_CellNuclei/HSV_CellNuclei_Hue.tiff',
    'Images/Datenvorbereitung_Images/CellNuclei/HSV_CellNuclei/HSV_CellNuclei_Saturation.tiff',
    'Images/Datenvorbereitung_Images/CellNuclei/HSV_CellNuclei/HSV_CellNuclei_Value.tiff',
    'Images/Datenvorbereitung_Images/YeastCells/HSV_Yeast/HSV_YeastCells_Hue.png',
    'Images/Datenvorbereitung_Images/YeastCells/HSV_Yeast/HSV_YeastCells_Saturation.png',
    'Images/Datenvorbereitung_Images/YeastCells/HSV_Yeast/HSV_YeastCells_Value.png',
    'Images/Datenvorbereitung_Images/YeastCells/HSV_Yeast/HSV_YeastCells.png'
]

for img_path in image_paths:
    print(f"Processed image: {img_path}")
    bild = mpimg.imread(img_path)

    clustertes_bild = skcluster_image(bild, n_clusters=2)

    # Normalisieren für saubere Darstellung
    img_norm = (clustertes_bild - clustertes_bild.min()) / (clustertes_bild.max() - clustertes_bild.min())

    # Prüfen, ob das Bild mehrkanalig (3 Kanäle) ist -> vollständiges HSV-Bild?
    if len(img_norm.shape) == 3 and img_norm.shape[2] == 3:
        # HSV -> RGB konvertieren (OpenCV erwartet Hue 0-179, hier evtl. skalieren)
        # Falls Bild float mit H in 0-1, auf OpenCV Bereich anpassen:
        hsv_for_cv = img_norm.copy()
        hsv_for_cv[..., 0] = hsv_for_cv[..., 0] * 179  # Hue auf 0-179 skalieren
        hsv_for_cv[..., 1] = hsv_for_cv[..., 1] * 255  # S auf 0-255 skalieren
        hsv_for_cv[..., 2] = hsv_for_cv[..., 2] * 255  # V auf 0-255 skalieren
        hsv_for_cv = hsv_for_cv.astype(np.uint8)

        rgb_img = cv2.cvtColor(hsv_for_cv, cv2.COLOR_HSV2RGB)

        plt.imshow(rgb_img)
        plt.axis('off')
        plt.show()

        # Zum Speichern als PNG im RGB-Format:
        save_name = os.path.splitext(os.path.basename(img_path))[0] + "_skclustered.png"
        #save_image(rgb_img, save_name)
        print(f"Saved as: {save_name}")

    else:
        # Einzelne Kanäle (Hue, Saturation oder Value), passende Colormap nutzen
        # Hue: cmap='hsv', sonst 'gray'
        cmap = 'hsv' if 'Hue' in img_path else 'gray'

        plt.imshow(img_norm, cmap=cmap)
        plt.axis('off')
        plt.show()

        # Für Speicherung in 8-Bit (uint8)
        clustered_uint8 = (img_norm * 255).astype(np.uint8)

        save_name = os.path.splitext(os.path.basename(img_path))[0] + "_skclustered.png"
        #save_image(clustered_uint8, save_name)
        print(f"Saved as: {save_name}")
```

# Creating Ground Truth with Cellpose a Deep Learning model

Reference: 
Stringer, C., Wang, T., Michaelos, M., & Pachitariu, M. (2021). Cellpose: a generalist algorithm for cellular segmentation. Nature Methods, 18(1), 100-106.

```python
#Cellpose on Yeast Cell image + saving 

import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from cellpose import models, plot

# save_image_gt-Funktion
def save_image_gt(img, name, ext="tiff"):
    downloads = os.path.join(os.path.expanduser("~"), "Downloads")
    filename = f"{name}.{ext.lstrip('.')}"
    output_path = os.path.join(downloads, filename)
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    try:
        imsave(output_path, img.astype("uint16"))
        print(f"✅ Bild gespeichert unter: {output_path}")
    except Exception as e:
        print(f"❌ Fehler beim Speichern von: {output_path}")
        print(e)

# 1. Bildpfad und Existenz prüfen
image_path = 'Original_Images/Clustering/Data/Yeast_Cells.jpg'  # <–– richtiges Suffix wählen
if not os.path.exists(image_path):
    raise FileNotFoundError(f"Datei nicht gefunden: {image_path}")
img = imread(image_path)

# 2. Modell laden und Segmentierung
model = models.CellposeModel(gpu=False, model_type='nuclei')
masks, flows, styles = model.eval(
    img,
    diameter=210,
    channels=[0, 0]
)

# 3. Nur die Maske speichern (Downloads-Ordner)
save_image_gt(masks, name="Yeast_Cells_Mask", ext="png")

# 4. Ergebnis anzeigen (optional)
fig = plt.figure(figsize=(8, 8))
plot.show_segmentation(fig, img, masks, flows[0], channels=[0, 0])
plt.tight_layout()
plt.show()
```

```python
#Cellpose (nuclei mode) on Cell Nuclei image + saving 

import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from cellpose import models, plot

# save_image_gt function (saves image to Downloads as uint16)
def save_image_gt(img, name, ext="tiff"):
    downloads = os.path.join(os.path.expanduser("~"), "Downloads")
    filename = f"{name}.{ext.lstrip('.')}"
    output_path = os.path.join(downloads, filename)
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    try:
        imsave(output_path, img.astype("uint16"))
        print(f"✅ Bild gespeichert unter: {output_path}")
    except Exception as e:
        print(f"❌ Fehler beim Speichern von: {output_path}")
        print(e)

# 1. Load image
image_path = 'Original_Images/Clustering/Data/Cell_Nuclei.tiff'
if not os.path.exists(image_path):
    raise FileNotFoundError(f"Datei nicht gefunden: {image_path}")
img = imread(image_path)

# 2. Extract only the blue channel (as 2D grayscale)
blue_channel = img[:, :, 2]

# Optional: Normalize to 0–255 range – if needed
blue_norm = (blue_channel / blue_channel.max() * 255).astype(np.uint8)

# 3. Load Cellpose model
model = models.CellposeModel(gpu=False, model_type='nuclei')

# 4. Run segmentation on blue channel (single channel, so channels=[0,0])
masks, flows, styles = model.eval(
    blue_norm,
    diameter=None,
    channels=[0, 0]
)

# 5. Save mask
save_image_gt(masks, name="Cell_Nuclei_Mask_BlueChannel", ext="png")

# 6. Show result
fig = plt.figure(figsize=(8, 8))
plot.show_segmentation(fig, blue_norm, masks, flows[0], channels=[0, 0])
plt.tight_layout()
plt.show()
```
