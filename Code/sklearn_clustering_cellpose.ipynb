# Creating "Ground Truth" / reference mask with Cellpose a Deep Learning model

Reference: 
Stringer, C., Wang, T., Michaelos, M., & Pachitariu, M. (2021). Cellpose: a generalist algorithm for cellular segmentation. Nature Methods, 18(1), 100-106.

```python
#Cellpose on Yeast Cell image + saving 

import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from cellpose import models, plot

# save_image_gt-Funktion
def save_image_gt(img, name, ext="tiff"):
    downloads = os.path.join(os.path.expanduser("~"), "Downloads")
    filename = f"{name}.{ext.lstrip('.')}"
    output_path = os.path.join(downloads, filename)
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    try:
        imsave(output_path, img.astype("uint16"))
        print(f"✅ Bild gespeichert unter: {output_path}")
    except Exception as e:
        print(f"❌ Fehler beim Speichern von: {output_path}")
        print(e)

# 1. Check image path and existence
image_path = 'Original_Images/Clustering/Data/Yeast_Cells.jpg'  
if not os.path.exists(image_path):
    raise FileNotFoundError(f"Datei nicht gefunden: {image_path}")
img = imread(image_path)

# 2. Load model and segmentation
model = models.CellposeModel(gpu=False, model_type='nuclei')
masks, flows, styles = model.eval(
    img,
    diameter=210,
    channels=[0, 0]
)

# 3. Save only the mask (Downloads folder)
save_image_gt(masks, name="Yeast_Cells_Mask", ext="png")

# 4. Display result
fig = plt.figure(figsize=(8, 8))
plot.show_segmentation(fig, img, masks, flows[0], channels=[0, 0])
plt.tight_layout()
plt.show()
```

```python
#Cellpose (nuclei mode) on Cell Nuclei image + saving 

import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from cellpose import models, plot

# save_image_gt function (saves image to Downloads as uint16)
def save_image_gt(img, name, ext="tiff"):
    downloads = os.path.join(os.path.expanduser("~"), "Downloads")
    filename = f"{name}.{ext.lstrip('.')}"
    output_path = os.path.join(downloads, filename)
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    try:
        imsave(output_path, img.astype("uint16"))
        print(f"✅ Bild gespeichert unter: {output_path}")
    except Exception as e:
        print(f"❌ Fehler beim Speichern von: {output_path}")
        print(e)

# 1. Load image
image_path = 'Original_Images/Clustering/Data/Cell_Nuclei.tiff'
if not os.path.exists(image_path):
    raise FileNotFoundError(f"Datei nicht gefunden: {image_path}")
img = imread(image_path)

# 2. Extract only the blue channel (as 2D grayscale)
blue_channel = img[:, :, 2]

# Optional: Normalize to 0–255 range – if needed
blue_norm = (blue_channel / blue_channel.max() * 255).astype(np.uint8)

# 3. Load Cellpose model
model = models.CellposeModel(gpu=False, model_type='nuclei')

# 4. Run segmentation on blue channel (single channel, so channels=[0,0])
masks, flows, styles = model.eval(
    blue_norm,
    diameter=None,
    channels=[0, 0]
)

# 5. Save mask
save_image_gt(masks, name="Cell_Nuclei_Mask_BlueChannel", ext="png")

# 6. Show result
fig = plt.figure(figsize=(8, 8))
plot.show_segmentation(fig, blue_norm, masks, flows[0], channels=[0, 0])
plt.tight_layout()
plt.show()
```
